```python
from random import choices
```

```python
my_words = choices(wordlist.english, k=12)
my_words
```

```python
secret_words = " ".join(my_words[:6])+"\n"+" ".join(my_words[6:])
print(secret_words)
```

```python
font_path =  "/usr/share/fonts/truetype/roboto/unhinted/RobotoCondensed-Regular.ttf"

black_img = Image.new("RGBA",(700,100),(0,0,0,255))

txt = Image.new('RGBA', black_img.size, 0 )

fnt = ImageFont.truetype(font_path,40)

d = ImageDraw.Draw(txt)
d.text((5,0), secret_words, (128,128,128,255) ,font=fnt)

img_result = Image.alpha_composite(black_img,txt).convert("L")
img_result.save("words.png")
img_result
```

```python
graph = base.SpectroGraphic("words.png",duration=5,min_freq=1000,max_freq=8000)
graph.save("words.wav")
```

```python
graph.play()
```
